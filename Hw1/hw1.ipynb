```python
#Assume 'tails' if 0<r<0.5, and 'heads' if 0.5<r<1

def simncoins(nsims,ncoins):
    result = np.zeros(nsims)
    for ind in range(nsims):
        result[ind] = np.sum(np.random.rand(ncoins)>0.5) #Total number of heads per simulation
    return result
```

```python
coins = np.array([2,4,16,32,64,128])
trials = 50000
for i,elem in enumerate(coins):
    plt.figure(i)  #New figure for each number of coins
    plt.hist(simncoins(trials,coins[i]),bins=coins[i]+1,range=(-0.5,coins[i]+0.5),alpha=0.6)
    plt.title('Coins = %i' %(coins[i]),fontsize=15)
    #plt.xlim(-0.5,coins[i]+0.5)
    plt.xlabel('Number of heads',fontsize=15)
    plt.ylabel('N (occurences)', fontsize=15)
    plt.savefig('SumitKumar_Coins{}.png'.format(str(coins[i])))
    
plt.show()
```
