## PROBLEM 2     

I'm looking at the probability of Tom Wolf's Success. The posterior probability distribution of Wolf's success based on a __uniform prior__ and a binomial likelihood is :-

$p(p|Wolf) = \dfrac{\mathcal{L}(Wolf|p)\ p(p)}{\int_0^1 \mathcal{L}(Wolf|p)\ p(p)\mathrm{d}p}$

where the binomial likelihood for N interviewees and M votes for Wolf is :-

$\mathcal{L}(Wolf|p) = \dfrac{N!}{M!(N-M)!}\ p^M\ (1-p)^{(N-M)}$ and $p(p) = 1$

The probability of Wolf's success within a certain % (say $s$) of the observed mean (i.e. 0.59) is :-

$\mathrm{Pr}(0.59 \pm s) = \int\limits_{0.59-s}^{0.59+s} p(p|Wolf) \mathrm{d}p$

```python
def binompost(n,m,lims): #Calculates the binomial posterior probability with the error limits given
    likelihood = comb(n,m)*(p**m)*((1-p)**(n-m))
    prior = 1.0 #Uniform prior 
    f = interp.interp1d(p,likelihood*prior,kind='cubic')
    norm,err_norm = integ.quad(f,0.0,1.0)
    numer,err_numer = integ.quad(f,(m/n)-lims,(m/n)+lims)
    plotpost(prior,likelihood,lims,numer/norm)
    return numer/norm
```

b)

__Bayes Theorem__: $p(B|A) = \dfrac{p(A|B) p(B)}{\sum p(A|B) p(B)}$

Suppose event __B__ = _Where I am (Pitt/Parents/Vac)_, and __A__ = _I can hear the Ocean_. I assume the priors remain the same as in the previous problem. 

At Pitt, I can't hear the ocean. So $p(A=Ocean|B=Pitt) \simeq 0$

At my Parents too, I can't hear the ocean. So $p(A=Ocean|B=Parents) \simeq 0$ 

At my vacation however, I can hear the ocean. Based on past vacations, I assume $p(A=Ocean|B=Vac) \simeq 0.6$

So, the probability of me being in __Pitt__ if I can hear the ocean is :-

$p(B=Pitt|A=Ocean) \propto p(A=Ocean|B=Pitt) = 0$

the probability of me being at my  __Parents__ if I can hear the ocean is :-

$p(B=Parents|A=Ocean) \propto p(A=Ocean|B=Parents) = 0$

the probability of me being at my  __Vacation__ if I can hear the ocean is :-

$p(B=Vac|A=Ocean) = \dfrac{p(A=Ocean|B=Vac)p(B=Vac)}{p(A=Ocean|B=Vac)p(B=Vac)} = \dfrac{0.6\times0.01}{0.6\times0.01} = 1$

__So based on my assumed priors and likelihoods, If I can hear the ocean, I'm certainly at vacation!__
