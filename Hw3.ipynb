```python
%pylab inline
import scipy.interpolate as interpol
import scipy.integrate as integrate
```

Probability of a single redshift being wrong = $p$

Assuming mismatch occurs if either object is wrong and using the axiom: $p(A\cup B) = p(A) + p(B) - p(A \cap B)$

$p_{mis} = 2p - p^2\      (0<p<1)$

The mismatch process can be regarded as Poisson, since the prior probability is so small ($\sim 0.005$). The likelihood is therefore :-

$\mathcal{L} (N_{tot}|N,\lambda) = \dfrac{\lambda^{N} e^{-\lambda}}{N!}$

$\lambda = N_{tot} p_{mis}$

$N_{tot} = 205$

While the priors are either uniform of Jeffeys :-

$p_U = 1/\lambda_{max} - \lambda_{min} = 1/205$

$p_J = \lambda^{-1/2}$

The posterior (e.g with uniform prior) :-

$post(\lambda|N) = \dfrac{\lambda^{N} e^{-\lambda}{p_U}}{\int \lambda^{N} e^{-\lambda}{p_U} \mathrm{d} \lambda}$

The probability that p<0.005 for N = 0,1 is :-

$Pr(N|p<0.005) = \int^{\lambda(0.005)}_0 post(\lambda|N) \mathrm{d}\lambda$

```python
ntot = 205.0
p_arr = np.linspace(0,1,1000)
#lam = ntot*p*(2-p)
lam_arr = ntot*p_arr*(2.0-p_arr)
mismatch = ntot*0.005*(2.0-0.005)

#priors
prior_u = 1.0/(lam_arr[-1]-lam_arr[0])
prior_j = lam_arr**(-0.5)

#Fixing the Jeffreys Prior by removing the value at infinity
whbad = np.where(np.isfinite(prior_j)==False)
prior_j[whbad]=0.0

def postprob(low,up,n,prior,ptype):
    if ptype == 'u':
        label = 'Uniform Prior'
    elif ptype == 'j':
        label = 'Jeffreys Prior'
    likelihood = (lam_arr**n)*(np.exp(-lam_arr))
    f = interpol.interp1d(lam_arr,likelihood*prior,kind='cubic')
    norm,err = integrate.quad(f,0,ntot)
    numer,err_num = integrate.quad(f,low,up) #integrating the posterior between 0 and lambda
    success = numer/norm
    plt.plot(lam_arr,likelihood*prior/norm,'r-',label=label)
    plt.xlabel(r'$\lambda$',fontsize=15)
    plt.ylabel('Posterior',fontsize=15)
    plt.axvline(up,ls='--',lw=1.4,color='k')
    plt.xlim(0,10)
    plt.legend()
    plt.show()
    print 'Probability of mismatch for p<0.005 = ', success
    print 'Probability of mismatch for p>0.005 = ', 1.0-success
    print 'Odds in favor of p<0.005 = ',success/(1.0-success)
```

```python
#Probability of p<0.005 and p>0.005 for N=0 with Uniform Prior

postprob(0,mismatch,0,prior_u,'u')
```

```python
#Probability of p<0.005 and p>0.005 for N=1 with Uniform Prior
postprob(0,mismatch,1,prior_u,'u')
```

```python
#Probability of p<0.005 and p>0.005 for N=0 with Jeffreys Prior
postprob(0,mismatch,0,prior_j,'j')
```

```python
#Probability of p<0.005 and p>0.005 for N=1 with Jeffreys Prior
postprob(0,mismatch,1,prior_j,'j')
```
