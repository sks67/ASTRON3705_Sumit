```python
nsims = np.arange(30)
coin1 = np.zeros_like(nsims)
coin2 = np.zeros_like(nsims)
points1 = np.zeros_like(coin1)
points2 = np.zeros_like(coin2)
for n in nsims:
    coin1[n] = np.random.rand()>0.5
    coin2[n] = np.random.rand()<0.5
    points1[n] = np.sum(coin1[0:n])
    points2[n] = np.sum(coin2[0:n])

plt.plot(nsims,points1,'r-o',nsims,points2,'b-o')
plt.xlabel('Trials')
plt.ylabel('Points')
plt.show()
    
```
